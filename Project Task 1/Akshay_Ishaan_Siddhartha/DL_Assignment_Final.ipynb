```python
# import libraries
import torch
import numpy as np
import torchvision
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import torchvision.datasets as datasets
from torchvision import transforms
from PIL import Image
from sklearn.decomposition import PCA
import pickle
```

```python
import os
transform = transforms.Compose([ #transform the images into tensors and normalize
    transforms.ToTensor(),  
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), 
])
class ImageDataset(Dataset):
    def __init__(self, img_dir, label_file, transform=None, target_transform=None):
        self.img_dir = img_dir
        with open(label_file, 'r') as f:
            self.labels = f.readlines()
        self.img_names = os.listdir(img_dir)
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_names[idx])
        image = Image.open(img_path)
        label = torch.tensor(float(self.labels[idx].strip()) - 1)

        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return image, label
```

```python
training_data = ImageDataset(img_dir = r'C:\Users\siddh\Downloads\Project Task 1\train_data',label_file=r'C:\Users\siddh\Downloads\Project Task 1\train_labels.txt', transform = transform)
train_loader = DataLoader(training_data, batch_size=4, shuffle=True)
val_data = ImageDataset(img_dir = r'C:\Users\siddh\Downloads\Project Task 1\val_data',label_file=r'C:\Users\siddh\Downloads\Project Task 1\val_labels.txt', transform = transform)
val_loader = DataLoader(val_data, batch_size=4, shuffle=True)
```

```python
features = []
labels = []

for images, lbls in train_loader:    
    images = images.view(images.size(0), -1).numpy()  
    features.append(images)
    labels.append(lbls.numpy())

features = np.concatenate(features, axis=0)
labels = np.concatenate(labels, axis=0)
#converted to numpy for PCA
```

```python

# Check if the PCA model file exists
if os.path.exists('pca_model.pkl'):
    # Load the PCA model from the file
    with open('pca_model.pkl', 'rb') as file:
        pca = pickle.load(file)
        features_pca_train = pca.fit_transform(features)

    print("PCA model loaded from pca_model.pkl")
else:
    # Create and fit the PCA model
    pca = PCA(n_components=0.95)
    features_pca_train = pca.fit_transform(features)
    print(f"PCA model created and fitted. Shape of transformed features: {features_pca_train.shape}")
    
    # Save the PCA model to the file
    with open('pca_model.pkl', 'wb') as file:
        pickle.dump(pca, file)
    print("PCA model saved to pca_model.pkl")
```

```python
# Convert back to torch tensors for training
features_pca_train = torch.tensor(features_pca_train, dtype=torch.float32)
labels = torch.tensor(labels, dtype=torch.long)

#pca dataset and dataloader
train_dataset_pca = torch.utils.data.TensorDataset(features_pca_train, labels)
train_loader_pca = torch.utils.data.DataLoader(train_dataset_pca, batch_size=64, shuffle=True)
```

```python
#apply pca on validation data as well

features = []
labels = []

for images, lbls in val_loader:
    
    images = images.view(images.size(0), -1).numpy()  
    features.append(images)
    labels.append(lbls.numpy())

features = np.concatenate(features, axis=0)
labels = np.concatenate(labels, axis=0)
```

```python
features_pca_val = pca.transform(features)
features_pca_val = torch.tensor(features_pca_val, dtype=torch.float32)
labels = torch.tensor(labels, dtype=torch.long)


val_dataset_pca = torch.utils.data.TensorDataset(features_pca_val, labels)
val_loader_pca = torch.utils.data.DataLoader(val_dataset_pca, batch_size=32, shuffle=True)
```

```python
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(1042,256)
        
        self.batch_norm1 = nn.BatchNorm1d(256)
        self.dropout1 = nn.Dropout(0.5)

        self.fc2 = nn.Linear(256, 256)
        self.batch_norm2 = nn.BatchNorm1d(256)
        self.dropout2 = nn.Dropout(0.5)

        self.fc3 = nn.Linear(256, 128)
        self.batch_norm3 = nn.BatchNorm1d(128)
        self.dropout3 = nn.Dropout(0.5)

        self.fc4 = nn.Linear(128, 64)
        self.batch_norm4 = nn.BatchNorm1d(64)
        self.dropout4 = nn.Dropout(0.5)

        self.fc5 = nn.Linear(64, 60)


    def forward(self, x):
        
        x = F.relu(self.batch_norm1(self.fc1(x)))
        x = self.dropout1(x)
        x = F.relu(self.batch_norm2(self.fc2(x)))
        x = self.dropout2(x)
        x = F.relu(self.batch_norm3(self.fc3(x)))
        x = self.dropout3(x)
        x = F.relu(self.batch_norm4(self.fc4(x)))
        x = self.dropout4(x)
        x = self.fc5(x)
        return x


# initialize the NN
model = Net()
print(model)
```

```python

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001,weight_decay=1e-4)
```

```python
num_epochs = 50  

for epoch in range(num_epochs):
    #training phase
    model.train()  
    running_loss = 0.0
    correct_train = 0
    total_train = 0
    i = 0
    
    for images, labels in train_loader_pca:
        
        optimizer.zero_grad()

        
        outputs = model(images)
        labels = labels.long()
        loss = criterion(outputs, labels)

        
        loss.backward()
        optimizer.step()

        
        running_loss += loss.item()

        # Calculate training accuracy
        _, predicted = torch.max(outputs.data, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()
        
    
    # Calculate training accuracy for this epoch
    train_accuracy = 100 * correct_train / total_train
    avg_train_loss = running_loss / len(train_loader_pca)

    # validation
    model.eval()  
    val_loss = 0.0
    correct_val = 0
    total_val = 0

    with torch.no_grad():  
        for images, labels in val_loader_pca:  
            outputs = model(images)
            labels = labels.long()
            loss = criterion(outputs, labels)

            
            val_loss += loss.item()

            
            _, predicted = torch.max(outputs.data, 1)
            total_val += labels.size(0)
            correct_val += (predicted == labels).sum().item()

    val_accuracy = 100 * correct_val / total_val
    avg_val_loss = val_loss / len(val_loader_pca)

    #show accuracy and loss
    print(f'Epoch [{epoch + 1}/{num_epochs}], '
          f'Train Loss: {avg_train_loss:.4f}, Train Accuracy: {train_accuracy:.2f}%, '
          f'Val Loss: {avg_val_loss:.4f}, Val Accuracy: {val_accuracy:.2f}%')

print('Training complete')
```

```python
def evaluate_model(model, val_loader):
    model.eval()  
    correct = 0
    total = 0

    with torch.no_grad():  
        for images, labels in val_loader:
            
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Validation Accuracy: {accuracy:.2f}%')


evaluate_model(model, val_loader_pca)
```
